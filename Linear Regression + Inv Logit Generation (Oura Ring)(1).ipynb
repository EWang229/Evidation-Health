```python
import pandas as pd
import numpy as np
oura_data = pd.read_csv("C:/Users/rocke/Downloads/archive/oura_2019_trends.csv")  # reading in csv
# oura_data = oura_data.to_numpy()
oura_data
```

```python
oura_data_filtered = oura_data.dropna()  # dropping na's if they appear in any row
oura_data_filtered
```

```python
from math import e
import random

stress_scores = []
for i in oura_data_filtered["sleep_score"]:
    beta = 0.01  # very small beta for the logistic regression (output point should look random w/ no apparent pattern)
    logit = e**(beta*i) / (1 + e**(beta*i))
    if random.random() < logit:
        stress_scores.append(0)
    else:
        stress_scores.append(1)
    
oura_invlogit = oura_data_filtered.iloc[:, 2:3]
oura_invlogit["stress_score"] = stress_scores
oura_invlogit
```

```python
stress_scores2 = []
for i in oura_data_filtered["sleep_score"]:
    beta = 3  # beta value for the linear regression
    noise = np.random.normal(0, 4, 1)  # adding in noise
    for j in noise:  # stripping the brackets from numpy array
        noise_val = j
    regression = beta*i + noise_val
    stress_scores2.append(regression)
    

oura_LR = oura_data_filtered.iloc[:, 2:3]
oura_LR["stress_score"] = stress_scores2
oura_LR
```

```python
oura_invlogit.to_csv(r'C:\Users\rocke\Downloads\oura_invlogit.csv')  # saving csv
oura_LR.to_csv(r'C:\Users\rocke\Downloads\oura_linear_regression.csv')  # saving as csv
```
